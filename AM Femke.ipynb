```python
import os
import numpy as np
import pandas as pd
#import holoviews as hv
import networkx as nx 
#from holoviews import opts
from bokeh.plotting import figure, show, output_file

#hv.extension('bokeh')

#from holoviews.element.graphs import layout_nodes

#only file0_dummy is used
file = 'GephiMatrix_author_similarity.csv'
file0_dummy = './nodeDummy.csv'
file1_dummy = './otherDummy.csv'
file2_dummy = './otherDummy2.csv'

#read csv file
df_data = pd.read_csv(file, sep=';', header=0, index_col=False)

#dit gebruiken ze in het voorbeeld om groepen aan te duiden, not sure whether to use or not
colormap = ["#444444", "#a6cee3", "#1f78b4", "#b2df8a", "#33a02c", "#fb9a99",
            "#e31a1c", "#fdbf6f", "#ff7f00", "#cab2d6", "#6a3d9a"] 

x = []
namesTo = []
namesFrom = []
namesWeight = []
hold = df_data.shape[0] #fixed number for column length

#loop that sets values first in lists for columns
i = 0
while i < hold:
    
        namesVal = list(df_data.columns.values)
        del namesVal[0]                  
        namesTimes = len(namesVal) - i
        namesHold = list(namesTimes* (df_data.iloc[i,0],))
        del namesHold[:1]
        
        #Tonames + delete names that are already linked
        x = namesVal
        del x[:(i+1)]
        
        #weights + delete weights that are already linked
        c = df_data.iloc[:,(i+1)].tolist()
        del c[:(i+1)]

        for ele in c: 
            if ele == 1: 
                c.remove(ele)
                
        namesFrom = list(namesFrom+namesHold)
        namesTo = list(namesTo+x)
        namesWeight = list(namesWeight+c)
              
        i += 1
        
#df from which the plot will be made
df_plot = pd.DataFrame(columns=['from', 'to', 'weight'])

#might still need to drop rows with namesHold 0
#puts said lists in columns
df_plot['from'] = namesFrom
df_plot['to'] = namesTo
#df_plot['weight'] = namesWeight

#delete edges with weight 0
df_plot = df_plot.loc[df_plot['weight'] != 0]

#transforms df in tuples (can we cut out df?) for plot
subset = df_plot[['from', 'to', 'weight']]
tuples = [tuple(x) for x in subset.values]
tuples

data=dict(
    xname=namesFrom,
    yname=namesTo
)

#not sure whether I need the following part
'''
#actual graph is made out of tuples
graph = hv.Graph(df_plot)
#layout of graph
layout_nodes(graph, layout=nx.layout.fruchterman_reingold_layout)
'''

#df_plot to check if values are correct
print(df_plot)

#plotting the figure
'''
p = figure(title="Author Similarity",
           x_axis_location="above",
           x_range= list(reversed(df_plot['from'].unique())), y_range=df_plot['from'].unique(),
           tooltips = [('names', '@namesFrom', '@namesTo')])

p.plot_width = 800
p.plot_height = 800
p.grid.grid_line_color = None
p.axis.axis_line_color = None
p.axis.major_tick_line_color = None
p.axis.major_label_text_font_size = "5pt"
p.axis.major_label_standoff = 0
p.xaxis.major_label_orientation = np.pi/3

p.rect('namesFrom', 'namesTo', 0.9, 0.9, source=df_plot,
       color='colors', alpha='alphas', line_color=None,
       hover_line_color='black', hover_color='colors')

output_file("auth_sim.html", title="auth_sim.py example")
show(p) # show the plot
'''
```
