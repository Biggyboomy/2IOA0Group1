```python
import os
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show, output_file
from bokeh.palettes import Plasma256 as palette
from bokeh.models import LinearColorMapper, ColorBar, NumeralTickFormatter


#only file0_dummy is used
file = './nodeDummy.csv'

#read csv file
df_data = pd.read_csv(file, sep=';', header=0, index_col=False)

#declaring lists
x = []
namesTo = []
namesFrom = []
namesWeight = []
hold = df_data.shape[0] #fixed number for column length

#loop that sets values first in lists for columns
i = 0
while i < hold:
    
        namesVal = list(df_data.columns.values)
        del namesVal[0]                  
        namesTimes = len(namesVal) 
        namesHold = list(namesTimes * (df_data.iloc[i,0],))
        

        #Tonames 
        x = namesVal

        
        #weights 
        c = df_data.iloc[:,(i+1)].tolist()
              
        namesFrom = list(namesFrom+namesHold)
        namesTo = list(namesTo+x)
        namesWeight = list(namesWeight+c)
              
        i += 1
        
#df from which the plot will be made
df_plot = pd.DataFrame(columns=['from', 'to', 'weight'])


#puts said lists in columns
df_plot['from'] = namesFrom
df_plot['to'] = namesTo
df_plot['weight'] = namesWeight

#dictionary for the data
data=dict(
    xname=namesFrom,
    yname=namesTo,
    weights = namesWeight
)


#colorbar layout
color_mapper = LinearColorMapper(low=df_plot['weight'].min(),
                                 high=df_plot['weight'].max(),
                                 palette=palette,
                                 id='cmpr')
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=5, border_line_color=None, location=(0,0))
color_bar.formatter = NumeralTickFormatter(format='0.0')



#plotting the figure
p = figure(title="Author Similarity", 
           output_backend="webgl",
           x_axis_location="above",
           x_range= list(reversed(df_plot['from'].unique())), y_range=(df_plot['from'].unique()),
           tooltips = [('names', '@yname, @xname'), ('weight', '@weights'), ])

#adding colorbar
p.add_layout(color_bar, 'right')

#creating the plot
p.plot_width = 750
p.plot_height = 700
p.grid.grid_line_color = None
p.axis.axis_line_color = None
p.axis.major_tick_line_color = None
p.axis.major_label_text_font_size = "6pt"
p.axis.major_label_standoff = 0
p.xaxis.major_label_orientation = np.pi/3

#putting the data in the grid
p.rect('xname', 'yname', 0.9, 0.9, source=data, fill_color={'field': 'weights','transform': color_mapper}, fill_alpha = 1, 
       alpha = 'weights', line_color=None)

show(p) # show the plot

```
