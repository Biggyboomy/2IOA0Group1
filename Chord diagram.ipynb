```python
import os, time
import numpy as np
import pandas as pd
import holoviews as hv
import networkx as nx
from holoviews import opts, dim
from pandas import DataFrame
from pathlib import Path



#start_time = time.time()

hv.extension('bokeh')
hv.output(size=200)

from holoviews.element.graphs import layout_nodes

#only file0_dummy is used
file0_dummy = './GephiMatrix_author_similarity.csv'
file1_dummy = './nodeDummy.csv'
file2_dummy = './dummy50.csv' 

#if output file exists, take csv file instead of going into while loop again
if not Path(file1_dummy[:-4] + '.out.csv').is_file():
    df_data = pd.read_csv(file1_dummy, sep=';', header=0, index_col=False)

    p = []
    d = []
    e = []
    f = []
    hold = df_data.shape[0]

    #loop that sets values first in lists for columns
    i = 0
    while i < hold:

            #Fromnames + delete row of names once listed
            b = list(df_data.columns.values)
            del b[0]                  
            a1 = len(b) - i
            a = list(a1* (df_data.iloc[i,0],))
            del a[:1]
            
            #Tonames + delete names that are already linked
            p = b
            del p[:(i+1)]

            #weights + delete weights that are already linked
            c = df_data.iloc[:,1].tolist()
            del c[:(i+1)]

            #remove people linked to themselves
            for ele in c: 
                if ele == 1: 
                    c.remove(ele)

            e = list(e+a)
            d = list(d+p)
            f = list(f+c)

            i += 1

    #df from which the plot will be made
    df_plot = pd.DataFrame(columns=['from', 'to', 'weight'])

    #puts said lists in columns
    df_plot['from'] = e
    df_plot['to'] = d
    df_plot['weight'] = f
    
    #delete edges with weight 0
    df_plot = df_plot.loc[df_plot['weight'] != 0.0]
    
    df_plot.to_csv(file1_dummy[:-4] + '.out.csv', sep = '\t', encoding = 'utf-8', index = False)
else:
    df_plot = pd.read_csv(file1_dummy[:-4] + '.out.csv', sep = '\t', encoding = 'utf-8', index_col=False)

    
chord = hv.Chord(df_plot)
chord.opts(
    opts.Chord(cmap='Category20', edge_cmap='Category20', edge_color=dim('weight').str(), 
               labels='name', node_color=dim('index').str()))


#print('Code segment took ' + str(time.time() - start_time))
```
